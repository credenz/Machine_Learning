```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```
from keras.applications import inception_v3

model = inception_v3.InceptionV3(weights='imagenet')

model.summary()
```

```
from PIL import Image

image = Image.open('ObjectDetection_InceptionV3/steel.jpg')

plt.imshow(image)
```

```

from keras import preprocessing

img = preprocessing.image.load_img('ObjectDetection_InceptionV3/steel.jpg', target_size=(299, 299))
x = preprocessing.image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = inception_v3.preprocess_input(x)
predictions = model.predict(x)
labels = inception_v3.decode_predictions(predictions, top=3)[0]

labels
```
